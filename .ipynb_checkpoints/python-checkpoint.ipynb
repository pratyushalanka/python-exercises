```python
from math import factorial as ft

n = int(input())

for row in range (n):
    for col_one in range (n-row):
        print(' ', end=' ')
        for col_two in range(row+1):
            res = ft(row ) / (ft(row - col_two) * ft(col_two))
            print(int(res), end=' ')
        print()

 
```
